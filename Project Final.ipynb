### **Observation**
* I have filled missing values in location as 'Not Specified' for missing data 

```python
import numpy as np

def fill_seniority(row):
    min_sal = row['salary_min']
    max_sal = row['salary_max']
    
    if pd.isna(row['seniority_level']):  # Only fill if missing
        
        if max_sal < 70000:
            return "Entry Level"
        
        elif 70000 <= max_sal < 120000:
            return "Mid Level"
        
        elif 120000 <= max_sal < 180000:
            return "Senior Level"
        
        elif min_sal >= 180000:
            return "Director / VP"
        
        else:
            return "Not Specified"
    
    return row['seniority_level']  # keep original if not null

data["seniority_level"] = data.apply(fill_seniority, axis=1)
```

### **Observation**
* I haved filled seniority_level into three categories like 'Entry level','Mid level','Senior level' according to salary_min and salary_max columns using .apply()

## **Checking outliers :**

### **Checking Numerical Columns**

```python
numerical_cols = data.select_dtypes(include=['int', 'float']).columns
numerical_cols
```

```python
print("\n === Outliers Count(IQR method) ===")

for col in numerical_cols:
    Q1= data[col].quantile(0.25)
    Q3= data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = data[(data[col] < lower) | (data[col] > upper)]
    print(f"{col} : {len(outliers)} outliers")
```

### **Observation**
* There are no outliers issues found in my dataset

##  **Categorical vs Categorical**

```python
top_status = data['Hiring_Status'].value_counts().head(10).index
top_types = data['Seniority_Level'].value_counts().head(10).index
filtered_data = data[
    data['Hiring_Status'].isin(top_status) & 
    data['Seniority_Level'].isin(top_types)
]
ct = pd.crosstab(
    filtered_data['Hiring_Status'],
    filtered_data['Seniority_Level'],
    normalize='index'
)
ct = ct.sort_index(axis=1)
plt.figure(figsize=(14,6))
ct.plot(kind='bar',  figsize=(14,6))

plt.xlabel("Hiring Status", fontsize=12)
plt.ylabel("Proportion", fontsize=12)
plt.title("Top 10 Hiring Status vs Top 10 Seniority Level", fontsize=15)

plt.xticks(rotation=45, ha='right')
plt.legend(title="Seniority Level", bbox_to_anchor=(1.05,1), loc='upper left')
plt.tight_layout()
plt.show()
```

### **Observation**
* Seniority levels such as Mid-Senior ,Entry Level and Associate appear most frequently in hiring, showing that companies hire more for experienced but not executive-level roles.

```python
top10 = data['Job_Title'].value_counts().head(10).index

plt.figure(figsize=(16,8))
sns.countplot(x='Job_Title', hue='Employment_Type', data=data[data['Job_Title'].isin(top10)])
plt.title('Top 10 Job Titles by Employment Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
